```python
import json
from tqdm import tqdm 

def read_datas(path, max_nums=None):
    cur_counts = 0
    with open(path, 'r', encoding='utf8') as f:
        for line in f:
            cur_counts += 1
            if max_nums and cur_counts > max_nums:
                return {}
            yield line

# 读取数据
def read_datas_am_thinking(path, data_cls):
    mul_ = 0; count = 0; datas = []
    with open('/code/zhaoxudong03/data_pipelines/datas/AM-DeepSeek-R1-0528-Distilled-{}.jsonl'.format(data_cls), 'w', encoding='utf8') as f_save:
        for line in tqdm(read_datas(path)):
            cur_ = json.loads(line)
            if len(cur_['conversations']) == 2:
                count += 1
                instruction = cur_['conversations'][0]['value']
                output = cur_['conversations'][1]['value']
                # print(cur_['conversations'][1].keys())
                think_, ans = cur_['conversations'][1]['info']['think_content'], cur_['conversations'][1]['info']['answer_content']
                # data_cls = cur_['conversations'][1]['info']['category']
                # think_ = think_.split('<think>')[1].split('</think>')[0].strip()
                # ans_ = ans.split('</answer>')[0]
                sample = {'prompt': instruction, 'think': think_, 'output': ans, 'data_cls': data_cls, 'source': 'AM-Thinking-v1-Distilled'}
                f_save.write(json.dumps(sample, ensure_ascii=False)+'\n')
            else:
                mul_ += 1
        print('多轮对话的数量为:::', mul_)
    
        # print(datas)
    # return datas
```

```python
# 汇总所有数据
path = '/code/zhaoxudong03/data_pipelines/datas'

import os 
import json 
from tqdm import tqdm 

def data_yield(path):
    with open(path, 'r', encoding='utf8') as f:
        for line in f:
            yield line
f_save = open('/code/zhaoxudong03/data_pipelines/training_data_zxd/merged_data.jsonl', 'w', encoding='utf8')

for file_name in os.listdir(path):
    if not file_name.endswith('.jsonl'):
        print(file_name + '---文件后缀名错误')
        continue
    count = 0; correct = 0; totals = 0
    file_path = os.path.join(path, file_name)
    for line in tqdm(data_yield(file_path)):
        totals += 1
        cur_ = json.loads(line)
        if 'prompt' in cur_ and 'source' in cur_:
            f_save.write(line+'\n')
            correct += 1
        else:
            print(file_name + '  数据字段有问题')
            count += 1
            break
    print('--------- {} 写入完成, 总计写入数据 {} 条, 原文件包含数据 {} 条-----------'.format(file_name, correct, totals))
```

```python
import os 
import json
import pandas as pd

path = '/code/zhaoxudong03/data_pipelines/training_data_zxd/step1/'

dfs = []
for name in os.listdir(path):
    file_path = os.path.join(path, name)
    with open(file_path, 'r', encoding='utf8') as f:
        for line in f.readlines():
            cur_ = json.loads(line)
            cur_new = {'source': cur_['source'], 'prompt': cur_['prompt'], 'clarity_score': cur_['score']['Clarity']['Score'], 'cg_score': cur_['score']['Cognitive_Complexity']['Score'], 'cls': name}
            dfs.append(cur_new)
```

```python
import pandas as pd

dfs1 = pd.DataFrame(dfs)
```

```python
test = dfs1.groupby(by=['cls'])
```
