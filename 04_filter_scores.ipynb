```python
import re
import os
import json
import pandas as pd
from tqdm import tqdm 

save_type = ['Code', 'Math', 'Reasoning_&_Logic', 'Data_Analysis', 'Creative_Writing', 'Instruction_Following_&_Text_Processing', 'Role_Playing_&_Chat', 'Humanities_Arts_&_Social_Sciences', 'Natural_&_Applied_Sciences', 'Business_&_Finance', 'Daily_Life_&_Health', 'General_Knowledge_QA']

def extract_json_from_string(text):
    """
    使用正则表达式从字符串中宽松地提取 JSON 字符串。
    这个正则表达式会查找以 '{' 开头，以 '}' 结尾，并且中间可以包含任何字符（包括换行符）的模式。
    """
    # 宽松的 JSON 匹配，考虑多行
    # re.DOTALL 使 '.' 匹配包括换行符在内的所有字符
    match = re.search(r'\{.*\}', text, re.DOTALL)
    if match:
        json_str = match.group(0)
        try:
            # 尝试解析提取到的字符串，确保它是有效的 JSON
            res = json.loads(json_str)
            return res
        except json.JSONDecodeError:
            # 如果不是有效的 JSON，则返回 None
            return None
    return None

def read_jsonl(path):
    results = []
    res = []; count = 0
    with open(path, 'r', encoding='utf8') as f:
        for line in tqdm(f.readlines()):
            count += 1
            cur_ = json.loads(line)
            if cur_['finish_reason'] != 'stop':
                continue
            try:
                cur_json = extract_json_from_string(cur_['output'])
                if 'Clarity' in cur_json and 'Cognitive_Complexity' in cur_json and 'Safety_Risk' in cur_json and 'Domain' in cur_json:
                    cur_['output'] = cur_json
                    res.append(cur_)
            except Exception as e:
                print('--------------------error-----------------', e)
                if count <= 30000:
                    print(cur_['output'])
                    break
                continue
        results = res
    print('data length :::', len(results), 'total length:::', count)
    return results
```

```python
# 第一步，按照数据的类别，分别保存文件，查看不同类别数量
def read_datas_(ori_paths):
    datas = []
    for path in ori_paths:
        cur_datas = read_jsonl(path)
        datas.extend(cur_datas)
    return datas

datas = read_datas_(['/code/zhaoxudong03/data_pipelines/training_data_zxd/wangfei_prompt_zh_03_30b_a3b_cls_before.jsonl', '/code/zhaoxudong03/data_pipelines/training_data_zxd/wangfei_prompt_zh_03_30b_a3b_cls_after.jsonl'])
```

```python
# 抽取每个类别中,8-10 分的数据
import re
import os
import json
import pandas as pd

save_type.append('Others')

def process_datas(k, values, v_clarity_down, v_cg_down, v_clarity_up=10, v_cg_up=10):
    # 筛选条件
    res = []
    for elem in values:
        # print(elem['score']['Clarity']['Score'])
        if  (elem['Clarity'] >= v_clarity_down and elem['Clarity'] <= v_clarity_up and elem['Cognitive_Complexity'] >= v_cg_down and elem['Cognitive_Complexity'] <= v_cg_up):
            res.append(elem)
    return res
        
def deal_datas_step2(datas, v_clarity_down=8, v_clarity_up=10, v_cg_down=5, v_cg_up=10):
    total_results = []
    others = []
    for k, v in datas.items():
        if k in save_type:
            # 处理数据，传入阈值
            res = process_datas(k, v, v_clarity_down, v_cg_down, v_clarity_up, v_cg_up)
            total_results.append({k: res})
        else:
            print('+++++++++++++++++++++++++++++++')
    #         others.extend(v)
    # res = process_datas('Other', others, v_clarity_down, v_cg_down)
    # total_results.append({'Other': res})
    return total_results

def step2_save_filter_datas_gai(data_type_all, v_clarity_down, v_cg_down, v_clarity_up=10, v_cg_up=10):
    # type_cur = list(data_type_all.keys())[0]
    str_name = 'clarity_over_{}_to_{}_and_cg_over_{}_to_{}'.format( v_clarity_down, v_clarity_up, v_cg_down, v_cg_up)
    res = deal_datas_step2(data_type_all, v_clarity_down=v_clarity_down, v_cg_down=v_cg_down, v_clarity_up=v_clarity_up, v_cg_up=v_cg_up)
    
    path = '/code/zhaoxudong03/data_pipelines/training_data_zxd/wangfei_zh/step2'
    save_path = path + '/' + str_name
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    df_dics = {}
    for elem in res:
        for k, v in elem.items():
            df_dics[k] = len(v)
            # print(k, len(v))
            save_path_ = os.path.join(save_path, "{}.jsonl".format(k))
            save_datas(save_path_, v)

    return df_dics

def save_datas(path, values):
    with open(path, 'w', encoding='utf8') as f_save:
        for value in values:
            f_save.write(json.dumps(value, ensure_ascii=False)+'\n')
    print('数据保存完成:::', path, len(values))
```
